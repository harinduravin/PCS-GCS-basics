```python
import ccdm
import numpy as np
from ccdm_initialize import initialize
```

```python
pOpt = np.array([0.537577302140556,0.322026673986155,0.115556317270981,0.024839706602308])
n = 256
[p_quant,num_info_bits,n_i] = initialize(pOpt,256)
```

```python
print(p_quant)
print(num_info_bits)
print(n_i)
```

```python
src_symbols = np.random.randint(0, 2, size=num_info_bits)
```

```python
code_symbols = ccdm.encode(src_symbols, n, n_i)
```

```python
print(code_symbols)
```

```python
len(code_symbols)
```

```python
print(src_symbols)
```

```python
decoded_bits = ccdm.decode(code_symbols, n_i, num_info_bits)
print(decoded_bits)
```

```python
np_src_symbols = np.array(src_symbols)
np_dec_bits = np.array(decoded_bits)
np.array_equal(np_src_symbols, np_dec_bits)
```
